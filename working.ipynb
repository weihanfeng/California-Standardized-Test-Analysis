```python
count = 0
count_not = 0
sat_codes = []
for i, v in enumerate(sat["CDCode"].astype(float).unique()):
    if v in districts["CDCode"].astype(float).unique() and v in act["CDCode"].unique():
        sat_codes.append(int(v))
        count +=1 
    else: 
        count_not +=1
```
